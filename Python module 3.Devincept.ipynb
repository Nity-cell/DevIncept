```python
import pandas as pd    
data = pd.read_csv("Latest Covid-19 India Status.csv") 
before = data.dtypes 
data["State/UTs"]= data["State/UTs"].astype(str) 
data["Active"]= data["Active"].astype(float)
after = data.dtypes 
print("BEFORE CONVERSION\n", before, "\n") 
print("AFTER CONVERSION\n", after, "\n") 
```
